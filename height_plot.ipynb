```python
NN_COLOR = '#e63946'
BNN_COLOR =  '#1d3557'
BNN_COLOR2 =  '#305890'
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(120, 220, 100)
fig, ax = plt.subplots(figsize = (10,5))
ax.plot(x, stats.norm.pdf(x, 160, 10), label = "Female height", color = NN_COLOR)
ax.plot(x, stats.norm.pdf(x, 180, 10), label = "Male height", color = BNN_COLOR)
ax.vlines(x = 170, ymin = 0, ymax = 0.04,
           colors = BNN_COLOR2,
           label = "Decision boundary",
          linestyles = "dotted")
ax.legend(loc = "upper left", fontsize = 14)
ax.set_xlabel("height", fontsize = 20)
ax.set_yticks([])
plt.show()
```
